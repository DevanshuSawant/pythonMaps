### Import libraries and get data from api

```python
import requests
import json
import pandas as pd
import geopandas as gpd
response_bus_locations = requests.get('http://tmtitsapi.locationtracker.com/api/getLastTrackingData')

json_data = json.loads(response_bus_locations.text)
bus_locations_df = pd.DataFrame(json_data['data'])
bus_locations_gdf = gpd.GeoDataFrame(bus_locations_df, geometry=gpd.points_from_xy(bus_locations_df.Longitude, bus_locations_df.Latitude))
bus_locations_gdf.set_crs(epsg=4326, inplace=True)
# print(bus_locations_df['VehId'])
bus_locations_gdf.explore()
```
